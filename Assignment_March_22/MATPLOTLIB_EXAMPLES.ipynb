__SIMPLE PLOTTING USING MATPLOTLIB__

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# scatter plot
```

```python
x = np.arange(0,20)
y = np.arange(30,50)
```

```python
plt.scatter(x,y, c = "g", marker = "*")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("SCATTER PLOT USING MATPLOTLIB")
plt.show()
```

```python
 #line graph
```

```python
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("LINE GRAPH USING MATPLOTLIB")
plt.show()
```

```python
df = [50,70,100,300,250,110,110,600,160,310]
```

```python
plt.boxplot(df)
plt.show()
```

```python
# BAR GRAPGH
```

```python
#COUNTRY
x = ["JAPAN", "INDIA", "CHINA", "AMERICA", "NORWAY"]
#NUMBER OF MEDALS IN OLYMPICS
y = [5,12,21,25,3]
```

```python
plt.bar(x,y)
plt.xlabel("COUNTRY")
plt.ylabel("MEDAL COUNT")
plt.title("BAR GRAPH USING MATPLOTLIB")
plt.show()
```

```python
# Pie chart
```

```python
x = [100,200,400,600]
labels = ['JAPAN','INDIA','AMERICA','RUSSIA']
colors = ['r','y','b','g']
```

```python
plt.pie(x, labels = labels, colors = colors, shadow = True)
plt.title("PIE CHART")
plt.show()
```

```python
# Exploding the pie chart
```

```python
e = (0.1,0.2,0.1,0.1)
plt.pie(x, explode = e, labels = labels, colors = colors, shadow = True)
plt.title("PIE CHART EXPLODED")
plt.show()
```

```python
# autopct 
```

```python
e = (0.1,0.2,0.1,0.1)
plt.pie(x, explode = e, labels = labels, colors = colors, autopct = "%1.2f%%", shadow = True)
plt.title("PIE CHART WITH autopct")
plt.show()
```
